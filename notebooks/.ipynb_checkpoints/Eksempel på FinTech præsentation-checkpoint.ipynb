### 1. Import of nessecary libraries 
Her ser man de biblioteker der skal bruges til at kunne k√∏re koden

```python
import numpy as np
import qiskit
```

### 2. Build the Quantum Circuit

```python
# Create a Quantum Register with 2 qubits.
qr = qiskit.QuantumRegister(2)

# Create a Classical Register with 2 bits.
cr = qiskit.ClassicalRegister(2)

# Create a Quantum Circuit acting on the q register
circuit = qiskit.QuantumCircuit(qr, cr)

# Prepare the draw funtion for the circuit
circuit.draw();
```

### 3. Add a H-gate to the circuit

```python
circuit.h(qr[0]);
```

### 4. Visualize the circuit

```python
# Visualizing the circuit using the draw function, which plots the circuit.
circuit.draw(output='mpl')
```

### 5. Simulating the Qubits using Qiskit Aer backend

```python
from qiskit.tools.visualization import plot_bloch_multivector
from qiskit import BasicAer

backend = BasicAer.get_backend('statevector_simulator')

run_job = qiskit.execute(circuit, backend)

result = run_job.result()

vector = result.get_statevector(circuit)

plot_bloch_multivector(vector)
```
