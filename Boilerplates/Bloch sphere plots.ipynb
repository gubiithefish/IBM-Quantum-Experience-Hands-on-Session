```python
import numpy as np
import qiskit
```

## Building the circuit

```python
# Create a Quantum Register with 2 qubits.
qr = qiskit.QuantumRegister(5)

# Create a Classical Register with 2 bits.
cr = qiskit.ClassicalRegister(5)

# Create a Quantum Circuit acting on the q register
circ = qiskit.QuantumCircuit(qr, cr)

# Prepate draw
circ.draw();
```

## Adding hadamard operation to the circuit

```python
# Add a H gate on qubit 0, putting this qubit in superposition.

circ.x(qr[1]);
circ.h(qr[0]);
circ.h(qr[0]);
circ.ccx(qr[0], qr[1],qr[2])
circ.x(qr[0]);
circ.h(qr[2]);
```

## Visualize Circuit

```python
# Visualizing the circuit using the draw function, which plots the circuit.
# Rendering of the circuit with a matplotlib drawing as found in many textbooks.
style = {'fontsize': 20, 'dpi': 300}

circ.draw(output='mpl', style=style)
```

# Simulating Circuits using Qiskit Aer
## Statevector backend

```python
# Import Aer
from qiskit import BasicAer

# Run the quantum circuit on a statevector simulator backend
backend = BasicAer.get_backend('statevector_simulator')

# Create a Quantum Program for execution
job = qiskit.execute(circ, backend)

# Get the results from the job
result = job.result()
```

```python
# Get the state vector, which is a complex vector of dimensions 2^n
outputstate = result.get_statevector(circ, decimals=3)

print(outputstate)
```

```python
from qiskit.tools.visualization import plot_bloch_multivector
plot_bloch_multivector(outputstate,"From 0 to 1")
```
